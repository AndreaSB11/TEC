```python
#Similitud con factor (1): Status de la relación sentimental
x=0
while x <=257:
    sentencias = pandas.read_csv('/Users/andreasb/Desktop/TEC/Concentración/Actividades/procesar_textos/sentencias_completas.csv')
    sentencias1 = sentencias.iloc[x,1]
    ejemplo = str(sentencias1)
    factor_1 = "miedo de denunciar al activo, nunca lo denunció por miedo,  refirió sentirse temorosa , podía buscarla y cumplir su amenaza de matarla, que ella no quería estar cerca de ese lugar, temor a que él pudiera encontrarla y ocasionarle un daño mayor, Trató de no mostrarle miedo ni que le estaba afectando porque si lo hacía la mataría, Tuvo miedo por la vida de su mamá, Regresaba con él, ya que le refería que le temía debido a que era una persona muy violenta, además de que la amedrentaba diciéndole que si lo dejaba, iba a atentar contra su familia, Le refirió que no se regresaba porque tenía miedo, La referida tenía miedo al agresor y a la violencia ejercida por este, pues dicho miedo inhibe y aleja a las afectadas, Los malos tratos hacía su hija nunca los denunció porque no la dejaba salir, que era muy celoso, que le decía qué necesidad tenía de andar en la calle, Al observar al acusado solicitó que no se advirtiera en la cámara porque le tenía miedo,Tenía miedo de contar lo sucedido por temor al acusado, reacción con temor al agresor, tuvo miedo hacia el denunciado, tenía miedo de que la matará o le hiciera algo a su hermano, ser el padre con apellidos, porque era la madre de sus hijos y su exesposa, por lo que su menor hijo empezó a llorar, Quien es hija de la víctima, No tenían hijos en común, Violencia familiar,  Sus hijos escuchaban normalmente las discusiones que tenían, Había agresiones físicas y verbales por parte de su padre a su mamá en otras ocasiones, La violencia reiterada que el menor vive, La menor se encontraba bajo violencia familiar.  La existencia de un ambiente de violencia familiar reiterada,  madre e hija se encontraban inmersas en un contexto de violencia y negligencia por parte del acusado."
    
    count_vect = CountVectorizer()
    Document1 = ejemplo
    Document2=factor_1
    
    corpus = [Document1,Document2]
    X_train_counts = count_vect.fit_transform(corpus)
    pd.DataFrame(X_train_counts.toarray(),columns=count_vect.get_feature_names(),index=['Document 1','Document 2'])
    
    vectorizer = TfidfVectorizer()
    trsfm=vectorizer.fit_transform(corpus)
    pd.DataFrame(trsfm.toarray(),columns=vectorizer.get_feature_names(),index=['Document 1','Document 2'])
    
    similaridad=cosine_similarity(trsfm[0:1], trsfm)
    print(similaridad)
    x=x+1
```
